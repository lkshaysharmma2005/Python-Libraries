```python
import numpy as np

arr = np.array([2, 3, 4, 6, 7, 8, 9, 12, 13, 16, 17, 23, 25, 27, 34, 37, 201])
```

```python
Q1 = np.percentile(arr,25)
Q3 = np.percentile(arr,75)
IQR = Q3 - Q1
IQR
```

```python
UF =  Q3 + 1.5*(IQR)
LF =  Q1 - 1.5*(IQR)
```

```python
UF
```

```python
LF
```

```python
list = []
for i in arr:
    if i>= LF and i<= UF:
        list.append(i)
arr2 = np.array(list)
```
