```python
import pandas as pd 
import seaborn as sns
from scipy.stats import f_oneway
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df = df[['age','pclass']].dropna()
```

```python
df['pclass'].unique()
```

```python
class_1 = df[df['pclass'] == 1]['age']
class_2 = df[df['pclass'] == 2]['age']
class_3 = df[df['pclass'] == 3]['age']
class_1
```

```python
f_stats , p_value = f_oneway(class_1,class_2,class_3)
```

```python
f_stats
```

```python
alpha = 0.05  ### if p_value is less than alpha then we reject the null hypothesis and vice versa

if p_value < alpha:
    print("I will reject Null hypothesis")
else:
    print("I will accept Null hypothesis")
```
