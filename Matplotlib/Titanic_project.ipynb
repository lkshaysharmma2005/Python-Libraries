## Step 2 - Step 2: Data Cleaning & Preprocessing Handle missing values:

age → fill with median or mean

embarked → fill with most frequent (mode)

Drop columns with too many missing values (deck often >70% missing)

Convert categorical to numeric:

sex → map to 0/1

Drop duplicate columns (who, alive) since they duplicate sex/survived

```python
df[df['age'].isna()]
```

```python
df["age"] = df["age"].fillna(df["age"].mean()).round(2)
df
```

```python
df.head()
```

```python
df['embarked']= df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
df
```

```python
df.dropna(thresh=int(0.3*len(df)),axis=1)
```

```python
df['sex'] = df['sex'].apply(lambda x: 1 if x == 'male' else 0)
df
```
