```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
plt.plot(x,y)
plt.title("Matplotlib Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x**2,y**2)
plt.subplot(2,2,4)
plt.plot(x**2,x**2)
```

```python
fig = plt.figure(figsize=(6,5), dpi=100) ## This makes an empty canvas
axis1 = fig.add_axes([0.1,0.1,1,1])  ## This adds a set of axes (a plotting area) to the canvas
axis1.plot(x,y)

axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

```python
fig = plt.figure(figsize=(6,5), dpi=100) ## This makes an empty canvas
axis = fig.add_axes([0.1,0.1,1,1])  ## This adds a set of axes (a plotting area) to the canvas
axis.plot(x,y)
plt.savefig("Matplotlib_save_plot.png")
```
