```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'Countries.csv')
```

```python
df
```

```python
## Which country has the highest population

df[df['population'] == df['population'].max()]
```

```python
## What is the capital of the country with the highest population

df[df['population'] == df['population'].max()]['capital_city']
```

```python
## Which country has the least population
df[df['population'] == df['population'].min()]
```

```python
df[df['population'] == df['population'].min()]['capital_city']
```

```python
## Top 5 countries with highest democratic score 	democracy_score
df[['country', 'democracy_score']].sort_values(by='democracy_score', ascending=False).head(5)
```

```python
## how many total regions are there
df['region'].nunique()
```

```python
unique_regions = df['region'].unique()
print(unique_regions)
```

```python
## how many countries lie in Eastern Europe region

df[df['region'] == "Eastern Europe"]["country"]
```

```python
## Worlds 2nd largest country political leader

df[df['population'] == df['population'].nlargest(2).iloc[1]]['political_leader']
```

```python
df[df['political_leader'].isna()]['country'].count()
```
